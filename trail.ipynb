```python
import spacy
from spacy import displacy

NER = spacy.load("en_core_web_sm")
```

```python
def contextToloc(text:str):
    text = NER(text)
    loc = []
    for word in text.ents:
        if word.label_ == "GPE":
            loc.append(word.text)
    return loc
```

```python
df["Locations"] = df["Context"].apply(contextToloc)
df
```

```python
def remove_dup(arr_input: list) ->list:
    unique_list = []
    for item in arr_input:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list
```

```python
df["Locations"] = df["Locations"].apply(remove_dup)
df
```

```python
def get_lat_long(location):
    try:
        # Get location information from Nominatim
        location_info = geolocator.geocode(location)
        if location_info:
            return location_info.latitude, location_info.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Error occurred: {e}")
        return None, None
```

```python
# Function to convert locations to latitude and longitude
def convert_to_lat_long(locations):
    latitudes = []
    longitudes = []
    for location in locations:
        lat, long = get_lat_long(location)
        latitudes.append(lat)
        longitudes.append(long)
    return latitudes, longitudes
```

```python
df[['Latitude', 'Longitude']] = df['Locations'].apply(lambda x: pd.Series(convert_to_lat_long(x)))
```
